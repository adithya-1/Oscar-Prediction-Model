```python
#Prediciton model for Best Picture
# importing required libraries
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv("../Data Set/Data For Prediction.csv")
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
train=df.drop(
    [
   #'year',
  #'certificate',
   # 'movie', 
                                                        # used to drop the columns not required i.e not important attributes 
                                                        # by not commenting the perticular attribute,
                                                        # initially all the attributes will be commented in order to get list of
                                                        # important attributes
    'genre', 
 'rate',
     'metascore', 
     #'votes',
     'gross', 
     'user_reviews',
     'critic_reviews',
     'popularity', 
     'awards_wins',
     #'awards_nominations',
 'Oscar_Best_Picture_won',
#'Oscar_Best_Picture_nominated', 
    'Oscar_Best_Director_won',
      
   #'Oscar_Best_Director_nominated',
    'Oscar_Best_Actor_won',
    #'Oscar_Best_Actor_nominated', 
    'Oscar_Best_Actress_won',
   #'Oscar_Best_Actress_nominated',
   #'other_bestPicture_won',
     #'other_director_won', 
    #'other_actor_won', 
   #'other_actress_won',
    'other_bestPicture_nominated', 
     'other_director_nominated',
       'other_actor_nominated',
    'other_actress_nominated'
    ],
     
     
     
     
     
    axis=1)
```

```python
test=pd.DataFrame()#initialising empty test dataframe
```

```python
test['Oscar_Best_Picture_won']=df['Oscar_Best_Picture_won'] # adding the output label to test the data
```

```python
X_train=[]
Y_train=[]
X_test=[]
Y_test=[]
for i in train.index:                                            #splitting the test data based on year 
                                                                 #train -from 2001-2016
                                                                 #test -2017
    if(int(train['year'][i])<2017):
        print(train.iloc[i])
        X_train.append(train.iloc[i])
        Y_train.append(test['Oscar_Best_Picture_won'][i])
    else:
        X_test.append(train.iloc[i])
        Y_test.append(test['Oscar_Best_Picture_won'][i])
        
```

```python
 # converting numpy array to dataframe
X_train=pd.DataFrame(X_train)
Y_train=pd.DataFrame(Y_train)
X_test=pd.DataFrame(X_test)
Y_test=pd.DataFrame(Y_test)
```

```python
movie=X_test['movie']
                                                            # dropping movie column to do ananlysis since Oscar winnings is least
                                                            # likely to depend on movie names 
X_train=X_train.drop(['movie'],axis=1)
X_test=X_test.drop(['movie'],axis=1)
```

```python
X_train=(X_train-X_train.min())/(X_train.max()-X_train.min())
Y_train=(Y_train-Y_train.min())/(Y_train.max()-Y_train.min()) #normalising the entire test train
X_test=(X_test-X_test.min())/(X_test.max()-X_test.min())
Y_test=(Y_test-Y_test.min())/(Y_test.max()-Y_test.min())
```

```python
X_train=X_train.drop(['year'],axis=1)#dropping year column
X_test=X_test.drop(['year'],axis=1)
```

```python
regr=LinearRegression()#training the linear regression model
```

```python
regr.fit(X_train,Y_train)
```

```python
pred=regr.predict(X_test)
```

```python
regr.score(X_test,Y_test)
```

```python
df_norm=pd.DataFrame(pred)

```

```python
df_norm=(df_norm-df_norm.min())/(df_norm.max()-df_norm.min())
```

```python
df_norm.sort_values(by=[0],ascending=False).head(5)  #getting top 5 values from prediction 
```

```python
 # converting  top 5 of these predictions to 
# probabilty using softmax and equating them to respective movie names
indexList=df_norm.sort_values(by=[0],ascending=False).head(5).index.to_list()
sort=df_norm.sort_values(by=[0],ascending=False).head(5)
movie=pd.DataFrame(movie.values)
deno=0
for i in indexList:
    deno+=np.exp(sort[0][i])
for i in indexList:
    print(movie[0][i],np.exp(sort[0][i])/deno,sep=':')
```
